```python
import numpy as np
import matplotlib.pyplot as plt
t = np.linspace(0, 2, 200)
signal_1 = np.sin(2 * np.pi * 3 * t)  # 3 Hz
signal_2 = np.sin(2 * np.pi * 7 * t)  # 7 Hz
#1
plt.figure(figsize=(12,6))
#2
plt.plot(t, signal_1,  
         color="r",
         linewidth=2,
         label="3 Hz"
        )  #don't forget the t!!!
#3
plt.plot(t, signal_2, color="b", linewidth=2, label="7 Hz") #it can also be written in a single line
#4
plt.xlabel("Time (s)")
#5
plt.ylabel("Amplitude")
#6
plt.title("Frequency Comparison")
#7
plt.legend(loc="upper right")
#8
plt.grid(True, 
         linestyle="--",
         alpha=0.3)
#9
plt.show()
```

```python
EXERCISE 3 â€” Subplots (Organization)
Objective: Use subplots to organize multiple visualizations.
# Create three signals
t = np.linspace(0, 1, 100)
y1 = np.sin(2 * np.pi * 5 * t)
y2 = np.cos(2 * np.pi * 5 * t)
y3 = y1 + y2  # Combined signal
# TASKS:
# 1. Create figure (15, 5) with 1 row, 3 columns
# 2. Subplot 1: plot y1, title "Sine Wave"
# 3. Subplot 2: plot y2, title "Cosine Wave"
# 4. Subplot 3: plot y3, title "Combined Signal"
# 5. All subplots: xlabel "Time (s)", ylabel "Amplitude", grid
# 6. Use plt.tight_layout() before showing
# 7. Display
```

```python
import numpy as np
import matplotlib.pyplot as plt
t = np.linspace(0, 1, 100)
y1 = np.sin(2 * np.pi * 5 * t)
y2 = np.cos(2 * np.pi * 5 * t)
y3 = y1 + y2  # Combined signal
#1
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
#2
axs[0].plot(t, y1,
          color="r",
          linewidth=2,
          label="Sine Wave"
         )
#3                                        #os labels nÃ£o sÃ£o necessÃ¡rios uma vez que existe apenas um sinal em cada grÃ¡fico.
axs[1].plot(t, y2,
          color="b",
          linewidth=2,
          label="Cosine Wave"
         )
#4
axs[2].plot(t, y3,
          color="black",
          linewidth=2,
          label="Combined Signal"
         )
#5
titles = ["Sine Wave", "Cosine Wave", "Combined Signal"]
for i in range(3):
    axs[i].set_xlabel("Time (s)")
    axs[i].set_ylabel("Amplitude")
    axs[i].set_title(titles[i])
    axs[i].grid(True, linestyle="--", alpha=0.3)
#6
plt.tight_layout() #avoids overlapping 
#7
plt.show()
```

```python
ExercÃ­cio 4 â€” Realistic ECG Simulation
Objective: Create a clinically-inspired signal visualization.
# Simulate ECG-like signal
fs = 500  # sampling rate
duration = 3  # seconds
t = np.arange(0, duration, 1/fs)

# Simple ECG approximation (P-QRS-T complex simulation)
ecg = np.zeros_like(t)
heart_rate = 75  # bpm
beat_interval = 60 / heart_rate  # seconds between beats

for beat_time in np.arange(0, duration, beat_interval):
    # QRS complex (sharp peak)
    qrs_idx = int(beat_time * fs)
    if qrs_idx < len(ecg) - 20:
        ecg[qrs_idx:qrs_idx+20] = np.concatenate([
            np.linspace(0, 1, 10),
            np.linspace(1, 0, 10)
        ])

# TASKS:
# 1. Create figure (14, 6)
# 2. Plot ECG signal: black line, width 1.5
# 3. xlabel: "Time (s)"
# 4. ylabel: "Voltage (mV)"
# 5. title: "Simulated ECG Signal (HR: 75 bpm)"
# 6. Grid with alpha 0.4
# 7. Set y-axis limits: (-0.2, 1.2)
# 8. Display
```

```python
fs = 500  # sampling rate
duration = 3  # seconds
t = np.arange(0, duration, 1/fs)

# Simple ECG approximation (P-QRS-T complex simulation)
ecg = np.zeros_like(t)
heart_rate = 75  # bpm
beat_interval = 60 / heart_rate  # seconds between beats

for beat_time in np.arange(0, duration, beat_interval):
    # QRS complex (sharp peak)
    qrs_idx = int(beat_time * fs)
    if qrs_idx < len(ecg) - 20:
        ecg[qrs_idx:qrs_idx+20] = np.concatenate([
            np.linspace(0, 1, 10),
            np.linspace(1, 0, 10)
        ])
#1
plt.figure(figsize=(14, 6))
#2
plt.plot(t, ecg,
         color="black",
         linewidth=1.5
        )
#3
plt.xlabel("Time (s)")
#4
plt.ylabel("Voltage (mV)")
#5
plt.title("Simulated ECG Signal (HR: 75 bpm)")
#6
plt.grid(True, linestyle="--", alpha=0.4)
#7
plt.ylim(-0.2, 1.2) #new function  plt.ylim(min, max)
#8
plt.show()
```

```python
ExercÃ­cio 5 â€” Signal + Noise Analysis
Objective: Visualize signal quality degradation.
# Clean signal
t = np.linspace(0, 2, 400)
clean_signal = np.sin(2 * np.pi * 5 * t)

# Noisy signal
noise = np.random.normal(0, 0.3, len(t))
noisy_signal = clean_signal + noise
# TASKS:
# 1. Create figure (14, 10) with 2 rows, 1 column
# 2. Top subplot:
#    - Plot clean_signal: green, width 2, label "Clean"
#    - Title: "Clean Signal"
#    - Grid, xlabel, ylabel, legend
# 3. Bottom subplot:
#    - Plot noisy_signal: red, width 1, alpha 0.7, label "Noisy"
#    - Plot clean_signal: green, width 2, alpha 0.5, label "Original"
#    - Title: "Signal with Noise (SNR degradation)"
#    - Grid, xlabel, ylabel, legend
# 4. tight_layout()
# 5. Display
```

```python
# Clean signal
t = np.linspace(0, 2, 400)
clean_signal = np.sin(2 * np.pi * 5 * t)

# Noisy signal
noise = np.random.normal(0, 0.3, len(t))
noisy_signal = clean_signal + noise
#1
fig, axs= plt.subplots(2, 1, figsize=(14, 10))
#2
axs[0].plot(t, clean_signal, color="green", linewidth=2, label="Clean")
axs[0].set_title("Clean Signal")
axs[0].grid(True, linestyle="--", alpha=0.3)
axs[0].set_xlabel("Time (s)")
axs[0].set_ylabel("Voltage (mV)")
axs[0].legend()          # it's like "print the label"
#3
axs[1].plot(t, noisy_signal, color="red", linewidth=1, alpha= 0.7, label="Noisy")
axs[1].plot(t, clean_signal, color="green", linewidth=2, alpha= 0.5, label="Original")
axs[1].set_title("Signal with Noise (SNR degradation)")
axs[1].grid(True, linestyle="--", alpha=0.3)
axs[1].set_xlabel("Time (s)")
axs[1].set_ylabel("Voltage (mV)")
axs[1].legend()          # it's like "print the label"
#4
plt.tight_layout()
#5
plt.show()
```

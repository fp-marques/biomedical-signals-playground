```python
EXERCISE 2 — Indexing & Slicing (Precision)
Objective: Access data surgically, without guessing.
import numpy as np
arr = np.array([5, 10, 15, 20, 25, 30, 35, 40, 45, 50])
# TASKS:
# 1. Access the 3rd element
# 2. Access the last 3 elements
# 3. Access elements at even indices (0, 2, 4, ...)
# 4. Reverse the array
# 5. Access elements from the 3rd to the second‑to‑last, stepping by 2


# Constraints:
# - Indexing / slicing only
# - No helper variables unless strictly necessary
```

```python
import numpy as np
arr = np.array([5, 10, 15, 20, 25, 30, 35, 40, 45, 50])
#1
print(arr[2])

#2
last_3 = arr[-3:]
print(last_3)

#3
even_i = arr[::2]
print(even_i)

#4
reverse = arr[::-1]
print(reverse)

#5
fun = arr[2:-1:2]
print(fun)
```

```python
EXERCISE 4 — Applied Data Analysis (Biomedical Context)
Objective: Use NumPy like a scientist, not a student.
Simulate body temperature measurements (°C)
1 measurement per hour, 24 hours
Mean = 37.0°C, Std = 0.3°C
# TASKS:
# 1. Create the temperature array using NumPy
# 2. Compute:
# - Maximum temperature
# - Minimum temperature
# - Mean temperature
# - Number of measurements > 37.5°C (fever threshold)


# Constraints:
# - Use NumPy statistical functions
# - Fever count MUST use boolean masking
```

```python
#1
import numpy as np
temp = np.random.normal(loc=37.0, scale=0.3, size=24)
#2
print(f"Max: {np.max(temp)}")
print(f"Min: {np.min(temp)}")
print(f"Mean: {np.mean(temp)}")
fever_count = np.sum(temp>37.5) #temp>37.5, gives an array of True´s / False´s
print(f"Fever Count: {fever_count}")
```

```python
EXERCISE 5 — Signal Generation (ECG Preparation)
Objective: Think in discrete time and signals.
# TASKS:
# 1. Create a time vector from 0 to 2 seconds with fs = 100 Hz
# (use np.arange, step = 1/fs)
# 2. Create a sinusoidal signal:
# y = sin(2π × 5 × t)
# Frequency = 5 Hz
# 3. Print:
# - Number of samples
# - Signal maximum
# - Signal minimum
# - Shape of t and y

# Constraints:
# - t and y must have identical shapes
# - No hard‑coded sample counts
```

```python
#1
import numpy as np
fs=100
f=5
step = 1/fs
vector= np.arange(0,2, step)

#2
y = np.sin(2 * np.pi * f * vector) #sin(2π × f × t)

#3
print(f"Number: {len(vector)}")
print(f"Max: {np.max(y):.2f}") #:.2f duas casas decimais em float
print(f"Min: {np.min(y):.2f}") 
print(f"Shape t: {vector.shape}")
print(f"Shape y: {y.shape}")
```
